```python
import pandas as pd
import os

os.makedirs('data', exist_ok=True)

df = pd.read_csv('train.csv.zip')
columns = df.columns
df['date'] = pd.to_datetime(df.created).dt.date
for d in sorted(df['date'].unique()):
    df[df['date']==d][columns].to_csv(f"train_{d}.csv", index=False, compression='zip')
    os.rename(f"train_{d}.csv", f"data/train_{d}.csv.zip")
del df

df = pd.read_csv('test.csv.zip')
columns = df.columns
df['date'] = pd.to_datetime(df.created).dt.date
for d in sorted(df['date'].unique()):
    df[df['date']==d][columns].to_csv(f"test_{d}.csv", index=True, compression='zip')
    os.rename(f"test_{d}.csv", f"data/test_{d}.csv.zip")
del df
```

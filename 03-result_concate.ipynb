```python
import pandas as pd
import os

from icecream import ic

dates = [
    '2021-07-05',
    '2021-07-06',
    '2021-07-07',
    '2021-07-08',
    '2021-07-09',
    '2021-07-10',
    '2021-07-11',
    '2021-08-02',
    '2021-08-03',
    '2021-08-04',
    '2021-08-05',
    '2021-08-06',
    '2021-08-07',
    '2021-08-08',
    '2021-09-15',
    '2021-09-16',
    '2021-09-17',
    '2021-09-18',
    '2021-09-19',
    '2021-09-20',
]

df = pd.DataFrame(columns=['1','2','3','4','5'])

for d in dates:
    ic()
    df = df.append(pd.read_csv(f"data/results_{d}.zip", index_col=0, compression='zip'), sort=False)

df_train_null_date = pd.read_csv("data/train_1970-01-01.csv.zip", compression='zip')
df_test_null_date = pd.read_csv("data/test_1970-01-01.csv.zip", index_col=0, compression='zip')

null_distr = df_train_null_date.Segment.value_counts(normalize=True).sort_index()
df_null_distr = pd.DataFrame([list(null_distr)]*len(df_test_null_date), columns = ['1','2','3','4','5'], index = df_test_null_date.index)

df = df.append(df_null_distr, sort=False)

df.fillna(0, inplace=True)
df.sort_index(inplace=True)
```

```python
df.to_csv('results.csv', index=False, float_format="%.5f", compression='zip')
os.rename("results.csv", "results.zip")
```
